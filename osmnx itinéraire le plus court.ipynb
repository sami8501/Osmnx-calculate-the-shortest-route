```python
import osmnx as ox
import networkx as nx
```

```python
place = 'Strasbourg'
place_query = {'city':'Strasbourg', 'state':'Strasbourg', 'country':'France'}
G = ox.graph_from_place(place_query, network_type='walk')
```

##### Sélectionnez un nœud d'origine et de destination et un cadre de sélection autour d'eux

```python
orig_node = ox.get_nearest_node(G, (48.614831, 7.787829))
dest_node = ox.get_nearest_node(G, (48.546784, 7.760704))
```

##### Calculer le chemin le plus court entre ces nœuds puis visualiser le résultat

```python
route = nx.shortest_path(G, orig_node, dest_node, weight='length', method='dijkstra')
fig, ax = ox.plot_graph_route(G, route, node_size=0, route_linewidth=5, figsize=(20,20), edge_color="w")
```

##### Calculons maintenant la distance (en mètres) de l'itinéraire entre ceux deux noeuds 

```python
net_dist = nx.shortest_path_length(G, orig_node, dest_node, weight='length',method='dijkstra')
net_dist
```
